Итоговый результат:

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

α = 0.95
β = 0.7
γ = 0.6
δ = 3.5
ε = 0.25
ρ = 0.1

def Aizawa(t, xyz):
    x, y, z = xyz
    dxdt = (z - β) * x - δ * y
    dydt = δ * x + (z - β) * y
    dzdt = γ + α * z - (z**3) / 3 - x**2 + (y**2) / (1 + ε * z) + ρ * z * x**3
    return [dxdt, dydt, dzdt]

initial = [1, 1, 1]  # Чувствуйте себя свободно записывая любой x, y и z!
start_t = (0, 50)  # Меняйте время как вы хотите!
Between_t = np.linspace(*start_t, 1000) # Поставьте сколько угодно вам количество точек!

solution = sp.integrate.solve_ivp(Aizawa, start_t, initial, t_eval=Between_t)

print("Первые 10 значений решения:")
for i in range(10):
    t = solution.t[i]
    x, y, z = solution.y[0][i], solution.y[1][i], solution.y[2][i]
    print(f"t = {t:.3f}, x = {x:.6f}, y = {y:.6f}, z = {z:.6f}")
print("Последние 10 значений решения:")
for i in range(-10, 0):
    t = solution.t[i]
    x, y, z = solution.y[0][i], solution.y[1][i], solution.y[2][i]
    print(f"t = {t:.3f}, x = {x:.6f}, y = {y:.6f}, z = {z:.6f}")

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(solution.y[0], solution.y[1], solution.y[2], lw=0.5)
plt.show()
```
